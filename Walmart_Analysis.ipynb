# Average unit price of all products sold

```python
average_price = df.groupby('product_name')['unit_price'].mean()
average_price
```

```python
plt.figure(figsize=(10,6))
bars = average_price.plot(kind = 'bar', color = 'skyblue')
plt.title('Average Unit Price of Products')
plt.xlabel('Product Name')
plt.ylabel('Average Unit Price')
plt.xticks(rotation = 45 ,ha='center')

for index, value in enumerate(average_price):
    plt.text(index, value, str(round(value,2)), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

# Which stores are underperforming in terms of sales volume ?

```python
sales_by_store  = df.groupby('store_id')['quantity_sold'].sum().sort_values()
sales_by_store
```

```python
threshold = sales_by_store.median()
underperforming_stores = sales_by_store[sales_by_store < threshold]
underperforming_stores
```

# What count of transaction involved promotions ?

```python
ax = sns.countplot(x = 'promotion_applied', data = df)
for bar in ax.containers:
  ax.bar_label(bar)
plt.title('Count of Transactions Involved in Promotions')
plt.xlabel('Promotion Applied')
plt.ylabel('Count')
```

# What is the total revenue generated from sales during promotional events ?

```python
df['revenue'] = df['quantity_sold'] * df['unit_price']
```

```python
promotional_revenue = df[df['promotion_applied'] == True]['revenue'].sum()
promotional_revenue
```

# Which product categories see the largest increase in demand during holidays ?

```python
holiday_sales = df[df['holiday_indicator'] == True]
holiday_demand = holiday_sales.groupby('product_name')['quantity_sold'].sum()
holiday_demand
```

```python
plt.figure(figsize=(12,6))
ax = holiday_demand.plot(kind = 'bar', color = 'skyblue', edgecolor='black')
for index, value in enumerate(holiday_demand):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10)
plt.title('Demand for Products during Holidays')
plt.xlabel('Product Name')
plt.ylabel('Demand')
plt.xticks(rotation = 45 ,ha='right',fontsize=10)
plt.grid(axis='y',linestyle='--',alpha=0.8)
plt.tight_layout()
plt.show()
```

# What is the correlation between weather conditions and sales performance ?

```python
a = df.groupby('weather_conditions')['revenue'].sum()
a
```

```python
plt.figure(figsize=(10,6))
ax = a.plot(kind = 'bar', color = 'skyblue', edgecolor='black')
for index, value in enumerate(a):
    plt.text(index, value, str(value), ha='center', va='bottom', fontsize=10)
plt.title('Revenue by Weather Conditions')
plt.xlabel('Weather Conditions')
plt.ylabel('Revenue')
plt.xticks(rotation = 45 ,ha='right',fontsize=10)
plt.grid(axis='y',linestyle='--',alpha=0.8)
plt.tight_layout()
plt.show()
```

# How does customer loyalty level affect purchasing patterns ?

```python
loyalty_analysis = df.groupby('customer_loyalty_level').agg({
    'revenue':['sum','mean'],
    'customer_id': 'nunique',
    'transaction_id': 'count'
}).reset_index()
loyalty_analysis_sorted = loyalty_analysis.sort_values(by = ('revenue','sum'), ascending = False)
loyalty_analysis_sorted
```

# Calculate the monthly sales and findout the trends

```python
df['transaction1_date'] = pd.to_datetime(df['transaction_date'])
```

```python
df['Month'] = df['transaction1_date'].dt.month_name()
```

```python
df['Month_Number'] = df['transaction1_date'].dt.month
```

```python
df['Month'].unique()
```

```python
monthly_revenue = df.groupby('Month')['revenue'].sum().reset_index()
plt.figure(figsize=(10,6))
plt.plot(monthly_revenue['Month'], monthly_revenue['revenue'], marker='o', linestyle='-', color='b')
plt.title('Monthly Sales Revenue')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.grid(True)
plt.show()
```

# Which product has the best performance in terms of volume sold ?

```python
best_product = df.groupby('product_name')['quantity_sold'].sum()
best_product
```

```python
plt.figure(figsize=(10,6))
best_product.plot(kind = 'bar', color = 'orange', edgecolor='black')

plt.title('Best Performing Products')
plt.xlabel('Product Name')
plt.ylabel('Quantity Sold')
plt.xticks(rotation = 45 ,ha='right',fontsize=10)
plt.grid(axis='y',linestyle='--',alpha=0.8)
plt.tight_layout()

for index, value in enumerate(best_product):
    plt.text(index, value+0.01 * max(best_product), str(value), ha='center', va='bottom', fontsize=10, color='black')
plt.show()
```
